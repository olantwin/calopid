# CNN for AdvSND neutrino flavour classification

```python
import tensorflow.keras
from tensorflow.keras.layers import (
    BatchNormalization,
    Concatenate,
    Conv2D,
    Dense,
    Dropout,
    Flatten,
    Input,
    MaxPooling2D,
    ReLU,
    RandomFlip,
)
from tensorflow.keras.models import Model
import tensorflow.keras.optimizers
import tensorflow.keras.metrics
import tensorflow.keras.losses
```

```python
from config import input_shape, input_shape_mf
```

```python
import tensorflow.keras.backend as K

K.set_image_data_format("channels_last")
```

```python
model_name = "CNN_incahuasi_logit_focus_nu_flavour_muvtau"
```

```python
drop_middle = 0.25

lr = 2e-4  # ; betaa1=30; betaa2=100; decay=1e-3

# AdvTarget

# TODO bigger kernels/windows in strip dimension (axis=1)?

target_h_input = Input(input_shape, name="target_h_in")
X_h = RandomFlip(mode="vertical")(target_h_input, training=True)
X_h = Conv2D(16, kernel_size=(1, 9), padding="same")(X_h)
X_h = BatchNormalization()(X_h)
X_h = ReLU()(X_h)
# X_h = CBAM()(X_h)
X_h = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_h)
X_h = Dropout(rate=drop_middle)(X_h)

X_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_h)
X_h = BatchNormalization()(X_h)
X_h = ReLU()(X_h)
# X_h = CBAM()(X_h)
X_h = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_h)
X_h = Dropout(rate=drop_middle)(X_h)

X_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_h)
X_h = BatchNormalization()(X_h)
X_h = ReLU()(X_h)
# X_h = CBAM()(X_h)
X_h = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_h)
X_h = Dropout(rate=drop_middle)(X_h)

X_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_h)
X_h = BatchNormalization()(X_h)
X_h = ReLU()(X_h)
# X_h = CBAM()(X_h)
X_h = MaxPooling2D(pool_size=(2, 2), padding="same")(X_h)

X_h = Flatten()(X_h)

target_v_input = Input(input_shape, name="target_v_in")
X_v = RandomFlip(mode="vertical")(target_v_input, training=True)
X_v = Conv2D(16, kernel_size=(1, 9), padding="same")(X_v)
X_v = BatchNormalization()(X_v)
X_v = ReLU()(X_v)
# X_v = CBAM()(X_v)
X_v = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_v)
X_v = Dropout(rate=drop_middle)(X_v)

X_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_v)
X_v = BatchNormalization()(X_v)
X_v = ReLU()(X_v)
# X_v = CBAM()(X_v)
X_v = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_v)
X_v = Dropout(rate=drop_middle)(X_v)

X_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_v)
X_v = BatchNormalization()(X_v)
X_v = ReLU()(X_v)
# X_v = CBAM()(X_v)
X_v = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_v)
X_v = Dropout(rate=drop_middle)(X_v)

X_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_v)
X_v = BatchNormalization()(X_v)
X_v = ReLU()(X_v)
# X_v = CBAM()(X_v)
X_v = MaxPooling2D(pool_size=(2, 2), padding="same")(X_v)

X_v = Flatten()(X_v)

# AdvTarget

mufilter_h_input = Input(input_shape_mf, name="mufilter_h_in")
X_mf_h = RandomFlip(mode="vertical")(mufilter_h_input, training=True)
X_mf_h = Conv2D(16, kernel_size=(1, 3), padding="same")(X_mf_h)
X_mf_h = BatchNormalization()(X_mf_h)
X_mf_h = ReLU()(X_mf_h)
# X_mf_h = CBAM()(X_mf_h)
X_mf_h = MaxPooling2D(pool_size=(1, 4), padding="valid")(X_mf_h)
X_mf_h = Dropout(rate=drop_middle)(X_mf_h)

X_mf_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_h)
X_mf_h = BatchNormalization()(X_mf_h)
X_mf_h = ReLU()(X_mf_h)
# X_mf_h = CBAM()(X_mf_h)
X_mf_h = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_mf_h)
X_mf_h = Dropout(rate=drop_middle)(X_mf_h)

X_mf_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_h)
X_mf_h = BatchNormalization()(X_mf_h)
X_mf_h = ReLU()(X_mf_h)
# X_mf_h = CBAM()(X_mf_h)
X_mf_h = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_mf_h)
X_mf_h = Dropout(rate=drop_middle)(X_mf_h)

X_mf_h = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_h)
X_mf_h = BatchNormalization()(X_mf_h)
X_mf_h = ReLU()(X_mf_h)
# X_mf_h = CBAM()(X_mf_h)
X_mf_h = MaxPooling2D(pool_size=(2, 2), padding="same")(X_mf_h)

X_mf_h = Flatten()(X_mf_h)

mufilter_v_input = Input(input_shape_mf, name="mufilter_v_in")
X_mf_v = RandomFlip(mode="vertical")(mufilter_v_input, training=True)
X_mf_v = Conv2D(16, kernel_size=(1, 3), padding="same")(X_mf_v)
X_mf_v = BatchNormalization()(X_mf_v)
X_mf_v = ReLU()(X_mf_v)
# X_mf_v = CBAM()(X_mf_v)
X_mf_v = MaxPooling2D(pool_size=(1, 4), padding="valid")(X_mf_v)
X_mf_v = Dropout(rate=drop_middle)(X_mf_v)

X_mf_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_v)
X_mf_v = BatchNormalization()(X_mf_v)
X_mf_v = ReLU()(X_mf_v)
# X_mf_v = CBAM()(X_mf_v)
X_mf_v = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_mf_v)
X_mf_v = Dropout(rate=drop_middle)(X_mf_v)

X_mf_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_v)
X_mf_v = BatchNormalization()(X_mf_v)
X_mf_v = ReLU()(X_mf_v)
# X_mf_v = CBAM()(X_mf_v)
X_mf_v = MaxPooling2D(pool_size=(2, 4), padding="valid")(X_mf_v)
X_mf_v = Dropout(rate=drop_middle)(X_mf_v)

X_mf_v = Conv2D(16, kernel_size=(3, 3), padding="same")(X_mf_v)
X_mf_v = BatchNormalization()(X_mf_v)
X_mf_v = ReLU()(X_mf_v)
# X_mf_v = CBAM()(X_mf_v)
X_mf_v = MaxPooling2D(pool_size=(2, 2), padding="same")(X_mf_v)

X_mf_v = Flatten()(X_mf_v)

X = Concatenate()([X_h, X_v, X_mf_h, X_mf_v])
X = Dense(4)(X)
X = BatchNormalization()(X)
X = ReLU()(X)
X = Dense(20)(X)
X = BatchNormalization()(X)
X = ReLU()(X)
X = Dropout(rate=0.2)(X)
# TODO Add dropout?
# X_ell = Dense(1)(X)
# X_had = Dense(1)(X)
X = Dense(1)(X)

model = Model(
    inputs=[target_h_input, target_v_input, mufilter_h_input, mufilter_v_input],
    outputs=X,
    name=model_name,
)

K.clear_session()

model.compile(
    optimizer="Adam",
    loss=tensorflow.keras.losses.BinaryFocalCrossentropy(from_logits=True),
    metrics=[
        tensorflow.keras.metrics.BinaryAccuracy(name="acc", threshold=0.5),
        tensorflow.keras.metrics.AUC(from_logits=True),
    ],
)
```

```python
model.save(f"{model_name}_e0.keras")
```
